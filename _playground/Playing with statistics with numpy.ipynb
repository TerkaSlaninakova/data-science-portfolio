# Playing with statistics with numpy
- illustrating the problem of skewing mean with an outlier
- calculating variance, std by hand
- probability density functions

### Illustrating the problem of skewing mean with an outlier

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data from normal distribution
mean=27000; std=15000; n_points=10000
incomes = np.random.normal(mean, std, n_points)

plt.hist(incomes, 50); plt.show()

print(f"mean: {np.mean(incomes)} | median: {np.median(incomes)}")
```

**Problem:** Mean can be skewed by outliers, median cannot

```python
incomes = np.append(incomes, [1000000000])
print(f"mean: {np.mean(incomes)} | median: {np.median(incomes)}")
```

### Calculating variance, std by hand

```python
small_dataset = np.random.randint(10, size=10)
print(small_dataset)
print(f"mean: {np.mean(small_dataset)}")
print(f"Diff. from the mean: {small_dataset - np.mean(small_dataset)}")
print(f"[Diff. from the mean]^2: {np.square(small_dataset - np.mean(small_dataset))}")
print(f"Avg from [Diff. from the mean]^2: {(np.square(small_dataset - np.mean(small_dataset))).sum() / 10}")

assert (np.square(small_dataset - np.mean(small_dataset))).sum() / 10 == small_dataset.var()

print(f"std [manually]: {np.sqrt((np.square(small_dataset - np.mean(small_dataset))).sum() / 10)}")
print(f"std: {small_dataset.std()}")
```

### Probability density functions (continuous) | probability mass functions (discrete)
- for a given range of values, gives me the probability of falling within a region of that range

```python
n_points = 100000
values = np.random.uniform(-10, 10, n_points)
plt.hist(values, 50)
plt.show()

exp_values = np.random.exponential(size=n_points)
plt.hist(exp_values, 50)
plt.show()

n, p = 10, 0.5
bin_values = np.random.binomial(n, p, size=n_points)
plt.hist(bin_values, 50)

plt.show()

poiss_values = np.random.poisson(size=n_points)
plt.hist(poiss_values, 50)

plt.show()
```

### Percentiles
- 50-th perc. is the median

### Moments
- measure the shape of a distribution
- 1st moment is just the mean
- 2nd moment is the variance
- 3rd moment is skew - "tail" on the left = negative skew, right = positive skew
- 4th moment is kurtosis - how sharp is the peak

```python
# Generate data from normal distribution
mean=27000; std=15000; n_points=10000
incomes = np.random.normal(mean, std, n_points)

assert np.median(incomes) == np.percentile(incomes, 50)

np.percentile(incomes, 50), np.percentile(incomes, 90)
```

```python
import scipy.stats as sp
sp.skew(incomes), sp.kurtosis(incomes)
```
