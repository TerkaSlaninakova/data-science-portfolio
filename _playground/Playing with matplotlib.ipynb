# Playing with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
x = np.arange(-3,3,0.001)

axes = plt.axes()
axes.set_xlim([-5, 5])
axes.set_ylim([0, 1.0])
axes.set_xticks(range(-5, 5))
axes.set_yticks(np.arange(0, 1, 0.1))

plt.plot(x, norm.pdf(x), c='b')
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:')

axes.grid()

plt.show()
```

```python
data = np.ones(100)
data[70:] += np.arange(30)

# lol :D
plt.xkcd()
plt.ylabel("My interest in\ngraph plotting", rotation=0, labelpad=50)
plt.xlabel("Time")

plt.xticks([]); plt.yticks([])

plt.annotate("Found out i can plot xkcd-styled graphs", xy=(70,1), arrowprops=dict(arrowstyle="->"), xytext=(30, 10))

plt.plot(data)
plt.show()
```

```python
# revert back xkcd
plt.rcdefaults()
v = range(1, 5, 1)
l = [str(i) for i in v]
c = ['r', 'b', 'g', 'c']

plt.pie(v, colors=c, labels=l)
plt.show()

plt.bar(v, l, color=c)
plt.show()
```
